```python
#6 -- celsius to fahrenheit
def toF(df):
    fah = []
    for i in range(len(df)):
        fah.append(df["Temp_C"][i])
        fah[i] = round(((df["Temp_C"][i])*(9/5)+32),2)
    return fah
```

```python
print(toF(df))
TF = toF(df)
df['Temp_F'] = TF
```

```python
#7 -- rename columns
df=df[['Date_Time', 'Temp_C', 'Temp_F','Dew_Point', 'Relative_Humidity', 'Wind_Speed', 'Visibility', 'Pressure', 'Weather']]
```

```python
#10 and 11 -- monthly f-temp and rearrange
month=df.Date_Time.dt.month
average=round(df.groupby(month).Temp_F.mean(),2)
monthly_avg=pd.DataFrame({'Temp_F':average})
monthly_avg
```
